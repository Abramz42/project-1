## Exploratory analysis and visualization

```python
datacopy.sample(3)
```

```python
y = datacopy.groupby('Continent')[['World Population Percentage']].sum()
plt.figure(figsize=(10,5))
plt.title('continents population percentage relationships')
plt.bar(y.index, y['World Population Percentage']);
```

```python
x = datacopy.groupby('Continent')[['2022 Population']].sum()
plt.figure(figsize=(10,5))
plt.title('2022 population relationship between continents')
plt.barh(x.index, x['2022 Population']);
```

```python
a = datacopy.groupby('Continent')[['Area (km²)']].sum()
a
```

```python
plt.pie(a['Area (km²)'], labels=a.index)
plt.legend(title='Continents', loc = 10);
```

```python
plt.subplot(2,1,1)
plt.bar(a.index, a['Area (km²)'])

plt.subplot(2,1,2)
plt.pie(a['Area (km²)'], labels=a.index)

plt.suptitle('Area occupied by each Continents');
```

```python
datacopy.sample(2)
```

```python
descendingdata = datacopy.sort_values('2022 Population',ascending = False)
descendingdata.head(10)
```

```python
firstdata = descendingdata.groupby('Continent').head(1)
firstdata
```

```python
plt.figure(figsize=(10,8))
plt.title('2022 Relationships of the most populated country in each continent')
plt.bar(firstdata['Country/Territory'],firstdata['2022 Population']);
```

```python
lastdata = descendingdata.groupby('Continent').tail(1)
lastdata
```

```python
plt.figure(figsize=(10,8))
plt.title('2022 Relationships of the least populated country in each continent')
plt.bar(lastdata['Country/Territory'],lastdata['2022 Population']);
```

## 7. what is the growth rate of each continents with the visualization?

```python
datacopy.sample(2)
```
